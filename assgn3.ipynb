```python
d = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}
df = pd.DataFrame(d)
print(df.shape)
print(df.info())
print(df.describe())
print(df.head(5))
print(df.tail(3))
print(df['Salary'].mean())
print(df['Bonus'].sum())
print(df['Age'].min())
print(df['Rating'].max())
df1 = df.sort_values(by='Salary', ascending=False)
print(df1)
df['Performance'] = pd.cut(df['Rating'], 
                           bins=[0, 4.0, 4.5, 5.0], 
                           labels=['Average', 'Good', 'Excellent'])
print(df[['Name', 'Performance']])
print(df.isnull().sum())
print(df.rename(columns={'Employee_ID': 'ID'}, inplace=True))
x=df[(df['Years_of_Experience'] > 5) & (df['Department'] == 'IT')]
print(x)
df['Tax'] = df['Salary'] * 0.10
print(df)
df.to_csv('modified.csv', index=False)
```
